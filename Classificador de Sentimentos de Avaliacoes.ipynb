```python
from transformers import pipeline

analisador = pipeline("sentiment-analysis", model="pysentimiento/bertweet-pt-sentiment")

# Entrada do usuário
avaliacao = input("Digite sua avaliação: ").title().strip()

resultado = analisador(avaliacao)[0]


print(avaliacao)
print(f"Sentimento: {resultado['label']}, Confiança: {resultado['score']:.2f}")
```
